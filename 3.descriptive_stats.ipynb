## 1. Introduction

Python Descriptive Statistics process describes the basic features of data in a study. It delivers summaries on the sample and the measures and does not use the data to learn about the population it represents.
Under descriptive statistics, fall two sets of properties- central tendency and dispersion. Python Central tendency characterizes one central value for the entire distribution. Measures under this include mean, median, and mode. Python Dispersion is the term for a practice that characterizes how apart the members of the distribution are from the center and from each other. Variance/Standard Deviation is one such measure of variability.

- **Mean** This function returns the arithmetic average of the data it operates on. If called on an empty container of data, it raises a StatisticsError.
- **Mode** This function returns the most common value in a set of data. This gives us a great idea of where the center lies.
- **Median** For data of odd length, this returns the middle item; for that of even length, it returns the average of the two middle items.

## 3. Measure of Dispersion
1. Range: The difference between the maximum and minimum values in a dataset.
2. Standard Deviation: A measure of the average deviation of each data point from the mean.
3. Variance: The square of the standard deviation, providing a measure of the spread of data.
4. Interquartile Range (IQR): The range between the first and third quartiles, less sensitive to outliers.
5. Mean Absolute Deviation (MAD): The average of the absolute differences between each data point and the mean.

These measures collectively provide a comprehensive view of how data points are dispersed in a dataset, taking into account different aspects such as the spread, average deviation, and resistance to outliers.

```python
# Computing the range for each feature
for i in df:
    try:
        print(f'{i} range {df[i].max() - df[i].min()}')
    except:
        pass
```

## 4. Measure of Shape

1. **Skewness**
Measures the symmetry of a dataset; a skewness of 0 indicates perfect symmetry. Negative skewness implies a longer tail on the left, while positive skewness indicates a longer tail on the right. Calculated using the third standardized moment.

2. **Kurtosis**
Quantifies the "tailedness" of a distribution. A kurtosis of 3 signifies a normal distribution. Leptokurtic distributions (kurtosis > 3) have heavier tails, while platykurtic distributions (kurtosis < 3) have lighter tails. Calculated using the fourth standardized moment.

High Kurtosis (Leptokurtic): A distribution with high kurtosis has heavy tails, indicating a higher probability of extreme values or outliers. In a leptokurtic distribution, the data points are more concentrated around the mean, and there is an increased likelihood of values far from the mean.

Low Kurtosis (Platykurtic): A distribution with low kurtosis has lighter tails than a normal distribution. In this case, there may be fewer extreme values or outliers, and the distribution is more spread out.

So, generally, high kurtosis suggests a higher likelihood of outliers, while low kurtosis suggests a distribution with fewer outliers. However, it's essential to note that kurtosis alone does not identify specific outliers; additional analysis and visualization techniques may be necessary to pinpoint and understand the nature of outliers in a dataset.
